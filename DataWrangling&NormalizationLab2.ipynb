```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url)
```

```python
df.head(5)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

df.columns = headers
```

```python
df.head(5)
```

```python
import numpy as np
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
# now that we have replace ? with NaN we need to evaluate 
#missing data to decide what to do with that. 
#Ultimately either delete the records or fill them with sth meaningful.
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print(" ")
```

```python
df.describe()
```

```python
avg_norm =df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized lossesï¼š ",avg_norm)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm, inplace=True)
```

```python
#now doublecheck if the Nan has been replaced with mean.
df["normalized-losses"].isnull()
```

```python
avg_bore =df["bore"].astype("float").mean(axis=0)
print("Average of boreï¼š ",avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)

df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)

df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)



```

```python
df['num-of-doors'].value_counts()
```

```python
#value_counts() shows most cars are four doors so we fill 4.
df['num-of-doors'].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#finally we delete rows without price data. 
#we want to predict price so missing price makes the entire row meaningless.

# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df1 = df.isnull()
df1.columns.value_counts()
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
# columns "city-mpg" and "highway-mpg" are represented by 
# mpg (miles per gallon). We now need to transform mgp into L/KM.
# formula conversion see below

# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

```python
df['highway-L/100km'] = 235/df["highway-mpg"]
```

```python
# rename column name from "highway-mpg" to "highway-L/100km"
# df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)

df.head()
```

```python
#now to normalize height width and length.

df['height'].max()

```

```python
# min-max scaling

df['height']= df['height']/df['height'].max() 
```

```python
# show the scaled columns
df[["length","width","height"]].head()
```

```python
df['width']= df['width']/df['width'].max() 
df['length']= df['length']/df['length'].max() 
```

```python
df[["length","width","height"]].head()
```

```python
#Binning in pandas

df["horsepower"]=df["horsepower"].astype(int, copy=True)

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")

bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins

group_names = ['Low', 'Medium', 'High']

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)

df["horsepower-binned"].value_counts()

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
